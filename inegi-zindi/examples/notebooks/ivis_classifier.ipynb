## Ivis Classifiers



```python
from ivis import Ivis
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
data = load_iris()
X = data.data
y = data.target
X = StandardScaler().fit_transform(X)

```

```python
ivis = Ivis(k=15, model='maaten', n_epochs_without_progress=5)
ivis.fit(X)
embeddings = ivis.transform(X)
embeddings.shape
```

```python
## visualization of embbedings


plt.figure(figsize=(8, 6))
plt.scatter(embeddings[:, 0], embeddings[:, 1], c=y, cmap='viridis', s=50)
plt.title('Ivis embeddings')
plt.show()
```

## IVIS Supervised 

```python
weight = 0.8
model = Ivis(k=15, model='maaten', n_epochs_without_progress=5)
model.fit(X, y)

embeddings = model.transform(X)
```

```python

plt.figure(figsize=(8, 6))
plt.scatter(embeddings[:, 0], embeddings[:, 1], c=y, cmap='viridis', s=50)
plt.title('Ivis embeddings Supervised')
plt.show()
```
